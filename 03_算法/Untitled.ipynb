### 素数筛

```python
import math
MAX_N = 10000
def get_prime(MAX_N):
    prime = [0 for _ in range(MAX_N + 5)]
    for i in range(2, int(math.sqrt(MAX_N + 1))):
        if prime[i] == 0:
            for j in range(2 * i, MAX_N + 1, i):
                prime[j] = 1
    for i in range(2, MAX_N + 1):
        if not prime[i]: yield i

# for i in get_prime(MAX_N):
#     print(i)
```

### 分解质因数

```python
def get_prime(MAX_N):
    prime = [0 for _ in range(MAX_N + 5)]
    for i in range(2, MAX_N + 1):
        for j in range(i, MAX_N + 1, i):
            if prime[j] == 0:
                prime[j] = i
    while True:
        num = int(input("请输入数字, 输入0退出:"))
        if not num: return
        s = num
        result = []
        while num > 1:
            result.append(str(prime[num]))
            num //= prime[num]
        print(f"{s} = {' * '.join(result)}")
```
